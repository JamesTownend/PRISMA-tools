# Reading HDF-EOS data

```python
# first of all install h5py library from command line with pip install h5py 
# enable h5py 
import h5py 

# open the PRISMA file 
f = h5py.File('/mnt/c/Users/townend/Documents/data/PRISMA/PRS_L2D_STD_20201002193710_20201002193714_0001.he5', 'r') 

# reading name and value for root attributes (metadata contained in HDF5 root) 
for attribute in f.attrs: 
    print(attribute,f.attrs[attribute]) 
```

```python
# reading names for all attributes (metadata) contained in HDF5 Groups 
# specific method for reading the values shall be built depending by the 
# specific metadata type (a single value, an array, a matrix, etc) 
def printname(name): 
    print(name) 
    
f.visit(printname) 
```

```python
import numpy as np
import pandas as pd 
import netCDF4
import xarray as xr
import h5py
```

```python
fname = '/mnt/c/Users/townend/Documents/data/PRISMA/PRS_L2D_STD_20201002193710_20201002193714_0001.he5'
group_name = 'HDFEOS/SWATHS/PRS_L2D_HCO/Data Fields/SWIR_Cube'
```

```python
ncf = netCDF4.Dataset(fname, diskless=True, persist=False)
```

```python
nch = ncf.groups.get('HDFEOS').groups.get('SWATHS').groups.get('PRS_L2D_HCO').groups.get('Data Fields')
ds = xr.open_dataset(xr.backends.NetCDF4DataStore(nch))
```

```python
ds = xr.open_mfdataset(fname, engine="h5netcdf", parallel=True)
```
